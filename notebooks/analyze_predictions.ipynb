```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import logging

# Configurar logging
logging.basicConfig(
    format='%(asctime)s - %(levelname)s - %(message)s',
    level=logging.INFO
)

# Cargar predicciones
predictions_path = Path("/Users/davidescudero/Documents/ITAM/PDT02/data/predictions")
latest_prediction = max(predictions_path.glob("predictions_*.csv"))
predictions = pd.read_csv(latest_prediction)

# Análisis básico
print("\nResumen de predicciones:")
print(predictions.describe())

print("\nPrimeras predicciones:")
print(predictions.head())

print("\nInformación del DataFrame:")
print(predictions.info())

# Visualización
plt.figure(figsize=(12, 6))
sns.histplot(data=predictions, x='item_cnt_month', bins=50)
plt.title('Distribución de Predicciones de Ventas')
plt.xlabel('Cantidad Predicha')
plt.ylabel('Frecuencia')
plt.show()

# Estadísticas adicionales
print("\nEstadísticas de las predicciones:")
print(f"Total de predicciones: {len(predictions):,}")
print(f"Predicciones > 0: {(predictions['item_cnt_month'] > 0).sum():,}")
print(f"Predicciones = 0: {(predictions['item_cnt_month'] == 0).sum():,}")
print(f"Media: {predictions['item_cnt_month'].mean():.2f}")
print(f"Mediana: {predictions['item_cnt_month'].median():.2f}")
print(f"Desv. Est.: {predictions['item_cnt_month'].std():.2f}")
```
